# Scatter Plots

````python
%matplotlib inline
import matplotlib.pyplot as plt

import numpy as np
np.random.seed(0)
````

````python
x = np.linspace(0, 10, 30)
y = np.sin(x)

plt.plot(x, y, 'o', color='black');
````

```python
matplotlib.pyplot.scatter(
    x, y,
    s=None,
    c=None,
    marker=None,
    ...,
    **kwargs
)
```

Marker:
- 'o'
- 'v'
- '^'
- '<'
- '>'
- '8'
- 's'
- 'p'
- '*'
- 'h'
- 'H'
- 'D'
- 'd'
- 'P'
- 'X'

````python
plt.scatter(x, y, marker='o');
````

````python
plt.scatter(x, y, marker='X');
````

````python
x = np.random.randn(100)
y = np.random.randn(100)
colors = np.random.rand(100)
area = 2*np.pi*y**2

#c,s,alpha could all be scalar (one value for all xy pairs) or a vector for every value

plt.scatter(
    x, y,
    c=colors,
    cmap='viridis',
    s=area,
    alpha=0.8
);
````

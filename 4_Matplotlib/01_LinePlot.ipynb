# Line Plots

```python
matplotlib.pyplot.plot(
    *args,
    scalex=True,
    scaley=True,
    data=None,
    ...,
    **kwargs
)
```

| Property                     | Description                                                  |
|------------------------------|--------------------------------------------------------------|
| alpha                        | scalar or None                                               |                                  
| color or c                   | color                                                        |
| dashes                       | sequence of floats (on/off ink in points) or (None, None)    |
| label                        | object                                                       |
| linestyle or ls              | {'-', '--', '-.', ':', '', (offset, on-off-seq), ...}        |
| linewidth or lw              | float                                                        |
| marker                       | marker style string, Path or MarkerStyle                     |                                              

````python
#jupyter magic command, shows plot figure in output instead of object
%matplotlib inline

import matplotlib.pyplot as plt

import numpy as np
np.random.seed(0)
````

### Figure and Axes

Matplotlib is strongly object oriented and its principal objects are the figure and the axes.

A figure is like a canvas. The canvas can be used in two ways, placing other objects on it and saving its contents with savefig.

An axes is sort of a Swiss Army knife. You can place many axes inside a figure using one of many different methods.

````python
fig = plt.figure()
ax = plt.axes()
````

````python
x = np.linspace(start=0, stop=2 * np.pi, num=50)

print(x)
````

````python
fig = plt.figure()
ax = plt.axes()

y = np.sin(x)

ax.plot(x, y);  #semicolon is also an jupyter magic commands, short form of plt.show()
````

````python
plt.plot(x, np.sin(x));
````

````python
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
plt.show()
````

## Adjusting the Plot: Line Colors and Styles

### Pre-Defined Colors

- b: blue
- g: green
- r: red
- c: cyan
- m: magenta
- y: yellow
- k: black
- w: white

````python
plt.plot(x, np.sin(x - 0.0 * np.pi), color='b');
````

````python
plt.plot(x, np.sin(x - 0.0 * np.pi), color='blue')
plt.plot(x, np.sin(x - 0.5 * np.pi), color='#FFDD44')
````

### Linestyles

- '-'
- '--'
- '-.'
- 'solid'
- 'dashed'
- 'dashdot'
- 'dotted'

````python
plt.plot(x, x + 0, linestyle='solid')
plt.plot(x, x + 1, linestyle='dashed')
plt.plot(x, x + 2, linestyle='dotted');
````

## Axes Limits

````python
plt.plot(x, np.sin(x))

plt.xlim(-1, 7)
plt.ylim(-1.25, 1.25);
````

````python
plt.plot(x, np.sin(x))
plt.axis([-1, 7, -1.25, 1.25]);
````

````python
plt.plot(x, np.sin(x))
plt.axis('tight');
````

````python
plt.plot(x, np.sin(x))
plt.axis('equal');
````

## Labeling Plots

````python
plt.plot(x, np.sin(x), linestyle="-.")
plt.plot(x, np.cos(x), linestyle="--")
plt.title("Curves")
plt.xlabel("x")
plt.ylabel("y")
plt.axis('tight')
plt.legend(["sin(x)", "cos(x)"]);
````
